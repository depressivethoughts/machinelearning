#### Полиноминальная регрессия

```python
# кодирование категориальных признаков
data_encoded = pd.get_dummies(data_cleaned, columns=['Car', 'Model'], drop_first=True)

scaler = StandardScaler()
numerical_features = ['Volume', 'Weight']
data_encoded[numerical_features] = scaler.fit_transform(data_encoded[numerical_features])

X = data_encoded.drop('CO2', axis=1)
y = data_encoded['CO2']

data_encoded
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score

# Создание полиномиальных признаков степени 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Разделение данных на обучающую и тестовую выборки
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

# Линейная регрессия с полиномиальными признаками
poly_reg = LinearRegression()
poly_reg.fit(X_train_poly, y_train)
y_pred_poly = poly_reg.predict(X_test_poly)

# Вычисление коэффициента детерминации (R²) для полиномиальной регрессии
r2_poly = r2_score(y_test, y_pred_poly)
print(f"Коэффициент детерминации (R^2): {r2_poly}")
```
